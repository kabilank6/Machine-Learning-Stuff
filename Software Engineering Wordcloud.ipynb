**Step 1: Preparing Data for Wordcloud**

```python
#read all the compulsory undergraduate course descriptions
with open("Software Engineering at UoA.txt", "r") as se:
    softeng_data = se.read()

#check contents
print("Our sample is: ", softeng_data[:200])

#import WordCloud which allows us to visually represent data later and also keeps track of stopwords
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

#generate wordcloud data
wordcloud = WordCloud(stopwords=stopwords, max_words=20, \
                     background_color="white").generate(softeng_data)
```

**Step 2: Displaying the Initial Wordcloud**

```python
import matplotlib.pyplot as mpLib
mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```

**Step 3: Enhancing Wordcloud**

```python
#add more words to ignore
stopwords.update(["students", "including", "introduction", "tools", "digital"])

#redo stop words. Limit number of words
wordcloud = WordCloud(stopwords=stopwords, max_words=10, \
                     background_color="azure").generate(softeng_data)
mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```
